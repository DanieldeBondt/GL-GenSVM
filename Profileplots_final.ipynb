```python
import numpy as np
from My_GenSVM import My_GenSVM
from sklearn.model_selection import train_test_split



```

```python
#loading own data
data_name = "breasttissue.csv"
data = np.loadtxt(data_name, delimiter=' ', skiprows=2)

starting_V = np.loadtxt("startV.csv", delimiter='\t')

x, y = np.hsplit(data, [-1])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

```

```python
# setting hyper parameters
roh = "unweighted"
kappa = 0
p = 1
epsilun = 10**-6
max_iter = 10**10
extension = True
lambdas = [2**4, 2**3, 2**2, 2, 1, 2**-1, 2**-2, 2**-3, 2**-4, 2**-5, 2**-6, 2**-7, 2**-8, 2**-9, 2**-10, 2**-11, 2**-12, 2**-13, 2**-14]
lambdas = lambdas[::-1]
# lambdas = [2**8]
norms = np.zeros((len(lambdas), x.shape[1]+1))
hitrates = np.zeros(len(lambdas))
testrates = np.zeros(len(lambdas))
aris = np.zeros(len(lambdas))
times = np.zeros(len(lambdas))


for testrun in range(len(lambdas)):
    SVM = My_GenSVM(x_train, y_train, rho=roh, lamb=lambdas[testrun], kappa=kappa, epsilun=epsilun, p=p, description=data_name, extension=extension,
                max_iter=max_iter, burn_in=51, seed=1212)
    SVM.print_model()
    solution = SVM.fit_im(starting_V, printing=False)
    optimal_V = solution[0]
    starting_V = optimal_V
    for rownumber in range(optimal_V.shape[0]):
        norms[testrun][rownumber] = np.linalg.norm(optimal_V[rownumber])
    testrates[testrun] = SVM.compute_hitrate(optimal_V, x_test, y_test)
    aris[testrun] = SVM.compute_ari(optimal_V, x_test, y_test)
    hitrates[testrun] = solution[2][-1]
    print("Lambda of "+ str(lambdas[testrun]) + " took time: "+ str(solution[3]))
    times[testrun] = solution[3]
    

np.savetxt("profile_norms_final_cv.csv", norms, delimiter=",")
np.savetxt("profile_hitrates_final_cv.csv", hitrates, delimiter=",")
np.savetxt("profile_testrates_final_cv.csv", testrates, delimiter=",")
np.savetxt("profile_times_final_cv.csv", times, delimiter=",")

    
        
    
    
```
